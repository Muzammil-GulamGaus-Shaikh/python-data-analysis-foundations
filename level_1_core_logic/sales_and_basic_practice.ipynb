🔹 Task 1: Sales Total Calculator

You are given a list:

sales = [1200, 1500, 1100, 1800, 2000]
You need to:

Calculate the total sales

Calculate the average sales

Find the maximum sale

Find the minimum sale

⚠️ Condition:

Do NOT use sum(), max(), or min()

Solve it manually using loops

```python
sales = [1200, 1500, 1100, 1800, 2000]
```

```python
#Calculate the total sales
total = 0
for i in sales:
    total = total + i
print(total)
```

```python
#Calculate the average sales
avg = 0
for i in sales:
    avg = total / len(sales)
print(avg)
```

```python
#Find the maximum sale
max_sale = sales[0]
for i in sales:
    if i > max_sale:
        max_sale = i
print(max_sale)
```

```python
#Find the minimum sale
min_sale = sales[0]
for i in sales:
    if i < min_sale:
        min_sale = i
print(min_sale)
```

🔹 Task 2: Even Number Filter (Data Cleaning Style)
numbers = [23, 44, 12, 67, 98, 31, 50]

Create a new list that contains:

Only even numbers

The same order as the original list

Expected output:

[44, 12, 98, 50]

```python
numbers = [23, 44, 12, 67, 98, 31, 50]
```

```python
even_num = []
for i in numbers:
    if i % 2 == 0:
        even_num.append(i)
print(even_num)
```

🔹 Task 3: Frequency Counter (Very Important 🔥)
data = ["apple", "banana", "apple", "orange", "banana", "apple"]

Create a dictionary that counts how many times each item appears.

Expected output:

{
  "apple": 3,
  "banana": 2,
  "orange": 1
}

Hint:

Use a dictionary

Use the concept of if key in dictionary

This is extremely important for real-world data analysis.

```python
data = ["apple", "banana", "apple", "orange", "banana", "apple"]
```

```python
frequency = {}
for i in data:
    if i in frequency:
        frequency[i] +=1
    else:
        frequency[i] = 1
print(frequency)
```

🔹 Task 4: Simple Data Cleaning
names = ["Ali", "  Ahmed", "Sara  ", "  Zoya  "]

Clean the data so the output becomes:

["Ali", "Ahmed", "Sara", "Zoya"]

Hint:

Use string methods

```python
names = ["Ali", "  Ahmed", "Sara  ", "  Zoya  "]
```

```python
clean_names = []
for i in names:
    clean_names.append(i.strip())
print(clean_names)
```
