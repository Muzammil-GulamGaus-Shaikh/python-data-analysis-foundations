🔹 Task 5: Transaction Filter transactions = [ {"id": 1, "amount": 500}, {"id": 2, "amount": 1500}, {"id": 3, "amount": 250}, {"id": 4, "amount": 2000} ]

Create a new list that contains:

Only transactions where amount > 1000

Expected output:

[ {"id": 2, "amount": 1500}, {"id": 4, "amount": 2000} ]


```python
transactions = [
    {"id": 1, "amount": 500},
    {"id": 2, "amount": 1500},
    {"id": 3, "amount": 250},
    {"id": 4, "amount": 2000}
]
```

```python
new_transaction = []
for i in transactions:
    if i["amount"] > 1000:
        new_transaction.append(i)
print(new_transaction)
```
