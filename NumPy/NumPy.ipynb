## Array Reshaping --> Reshape, Ravel, Flatten

```python
# Reshape
arr = np.array([1,2,3,4,5,6])
reshaped = arr.reshape(2,3)
reshaped
```

```python
reshaped2 = reshaped.reshape(3,2)
reshaped2
```

```python
# ravel --> convert 1D array
ravel = reshaped2.ravel()
ravel
```

```python
ravel[0] = 100 # Changes reflected into original array
ravel
reshaped2
```

```python
# flatten --> to 1D array
# return a copy of the array

flat = reshaped2.flatten()
flat
```

```python
flat[0] = 455
flat
```

```python
reshaped2
```

### Index Arrays --> Advanced Indexing

```python
#np.take --> built in func to perform indexing and sclicing
arr = np.array([10,20,30,40,50])
ind = [0, 2]
print(np.take(arr, ind))
```

```python
# Iterating with nditer()
arr = np.array([[1,2],[3,4]])
for x in np.nditer(arr):
    print(x, end = " ")
```

```python
# ndenumerate() -> both index + vlaue
for ind, x in np.ndenumerate(arr):
    print(ind, x)
```

### Cumulative Opeations

```python
# running total (cumsum)
arr # array([1, 2, 3])
print(np.cumsum(arr)) # 1 -> 1, 1+2 -> 3, 3+3 -> 6
print(np.cumprod(arr)) # 1 -> 1, 1*2 -> 2, 2*3 -> 6
```
