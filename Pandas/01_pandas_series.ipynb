### Indexing in Series

```python
# indexing
s[0]
```

```python
s[0:3] # start:stop(exclude):step (value to jump)
```

```python
s[0:5:2]# start:stop(exclude):step (value to jump)
```

### iloc --> Location based indexing

```python
s.iloc[3]
```

```python
s.iloc[[1,3,4]]
```

```python
index = ["apple", "banana", "grapes", "orange", "strawberry"]
```

```python
s.index = index
s
```

```python
s['grapes']
```

```python
# we can not use iloc in tihs index
# we can use iloc in original index

#s.iloc["grapes"] ---> gives an ERROR
s.iloc[3] #---> where as is works fine with ogr index
```

### loc ---> Lable based indexing

```python
s.loc['grapes']
```

```python
s.loc[['orange', 'strawberry']]
```

```python
# in lable based indexing your start as well as stop value are included in the output.
s['banana':'orange']
```

```python
fruit_protein = {
    "Avocado": 2.0,      # grams of protein
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}
```

```python
s2 = pd.Series(fruit_protein, name = "protein") # creating series from Dictionary
s2
```

```python
s2.iloc[0]
```

```python
s2.loc['Oranges']
```

```python
s2.loc[['Guava', 'Kiwi']]
```

### Conditional Selection

```python
s2>1
```

```python
s2[s2>1]
```

### logical operator

```python
# and 
s2[(s2>0.5) & (s2 <= 2)]
```

```python
# or
s2[(s2>0.5) | (s2 <= 2)]
```

```python
# not
s2[~(s2>1)]
```

### Modifing the Series

```python
s2["Mango"] = 2.8
```
